```python
# @title Ejecutar si la imagen es demaciado grande
imagen = "q.jpg" # @param {type:"string"}
factor_de_reduccion = 2 # @param {type:"slider", min:2, max:4, step:1}
imagen = f"ESRGAN/LR/{imagen}"



import PIL
import os
img=PIL.Image.open(imagen)
tam=img.size
nt=(tam[0]//factor_de_reduccion,tam[1]//factor_de_reduccion)
print(f"{tam}->{nt}")
img=img.resize(nt)
os.remove(imagen)
img.save(imagen)
```
