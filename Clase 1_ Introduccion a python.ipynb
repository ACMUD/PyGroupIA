#Condicionales
Un condicional es una estructura que requiere un booleano para tomar una decicion.


<center>
  
</center>


Opcionalemente se pueden agregar mas elementos a la estructura, como el `else` o el `elif`




#Funciones

Las funciones son segmentos de código que pueden llamarse en cualquier momento, estas pueden o no retornar algun valor

<center>
  
</center>

```python
def <nombre>(<parametros>):
  <codigo...>
```
Pueden devolder valores agregando al final la palabra reservada `return` y seguido aquello que retornará

```python
def <nombre>(<parametros>):
  <codigo...>
  return <variable o valor de retorno>
```

````python
#En este ejemplo se utilizaron ciclos, sin embargo eso no es lo importante sino la aplicación de la funcion para reutilizar un segmento de código.
def ord(list_:list):
  if (type(list_)==type([1,2,3])):
    for i in range(len(list_)):
      for j in range(i,len(list_)):
        if list_[i]>list_[j]:
          temp=list_[i]
          list_[i]=list_[j]
          list_[j]=temp
    return list_

a=[1,5,61,2,6,2,3,7,7]
print(ord(a))
b=[1,5,61,3,6,7,3,6,2]

print(ord(b))
````

#Funciones incognitas o lambda

A parte de las funciones como segmentos de codigo tambien hay funciones que solo ejecutan una linea de codigo y que normalmente solo se utilizan para guardar funciones matematicas, estas son funciones lambda



```python
<Variable>=lambda <parametro 1>,<parametro 2>,...,<parametro n>: <linea de codigo>
```



````python
g=lambda x,y: pow(x**2+y**2,1/2)

#evaluando la función
for i in range(-10,10,10):
  for j in range(-10,10,10):
    print("g({},{})={}".format(i,j,g(i,j)))
````

#try-except
Hay ocaciones que alguna operación no se puede cumplir porque genera un error, esto significa que el programa se detiene sin completar el proceso, para evitar esto se puede usar try para manejar estos errores como excepciones


```python
try:
  <codigo...>#Se ejecutara si no ocurre ningun error
except:
  <codigo...>#Se ejecuta si ocurre un error
```



````python
a=None
try:
  a=open("archivo.txt","r")
  print("No hay error")
except:
  a=open("archivo.txt","w")
  a.write("contenido")
  a.close()
  a=open("archivo.txt","r")
  print("Hubo error")
finally:
  print(a.read())
````

Los ciclos son estructuras que se repiten segun una condición

Untitled Diagram (2).png

Hay dos tipos de ciclos


Ciclo while, se repetira mientras se cumpla la condición
```python
while <Condicion>:
  <codigo...>
```

ciclo for, este tiene un funcionamiento diferente, puede iterar con una lista o con un rango, se detendrá antes de el ultimo valor del rango, es decir que si se coloca 10 terminará en 9
```python
for i in range(<numero de repeticiones>):
  <codigo...>
```

Tambien aceptan las sentencias `break` y `continue`, la primera detendrá el ciclo y el segundo hará que pase al siguiente paso del ciclo.

**Ejemplo 1**

factorial

$$n!=\prod_{m=1}^{n}m$$

Esta definición de factorial puede dar la idea de hacer un algoritmo iterativo


Untitled Diagram (4).png

````python
def factorial(a):
  valor=1
  for i in range(1,a+1):
    valor*=i
  return valor

i=0
while i<12:
  print("{:2}! = {}".format(i,factorial(i)))
  i+=1
````

#Recursividad

Una funcion recursiva es una funcion que puede definirse a si misma, un ejemplo de ello es el factorial cuya definicion se da con el mismo
$$n!=n(n-1)!$$


Untitled Diagram (5).png

Como puede observarse **no se definen variables ni ciclos a excepcion del parámetro**


**Ejemplo 1**

````python
factorial=lambda n: (n*factorial(n-1) if n>0 else 1)

for i in range(12):
  print("{:2} != {}".format(i,factorial(i)))
````

**Ejemplo 2**

````python
def fibonachi(n,a=0,b=1):
  if n>0:
    return a+fibonachi(n-1,b,a+b)
  else:
    return 1

for i in range(12):
  print("fibonachi({}) = {}".format(i,fibonachi(i)))
````
