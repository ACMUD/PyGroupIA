#Condicionales
Un condicional es una estructura que requiere un booleano para tomar una decicion.


<center>
  
</center>

```python
if(<condicion>):
  #se ejecuta si la condicion es verdadera
  <codigo...>
```
Opcionalemente se pueden agregar mas elementos a la estructura, como el `else` o el `elif`



```python
if(<condicion1>):
  #se ejecuta si la condicion es verdadera
  <codigo...>
elif (<condicion2>):
  #se ejecuta si la condicion es verdadera y la anterior no se cumplio
  <codigo...>
else:
  #se ejecuta si las anteriores condiciones no se cumplieron
  <codigo...>
```

````python
x=-10

if (x>0):
  print("x es positivo")
elif (x<0):
  print("x es negativo")
else:
  print("x es 0")
````

#Funciones

Las funciones son segmentos de código que pueden llamarse en cualquier momento, estas pueden o no retornar algun valor

<center>
  
</center>

```python
def <nombre>(<parametros>):
  <codigo...>
```
Pueden devolder valores agregando al final la palabra reservada `return` y seguido aquello que retornará

```python
def <nombre>(<parametros>):
  <codigo...>
  return <variable o valor de retorno>
```

````python
def implicacion(a,b):
  return (not a or b)

print(implicacion(True, True))
print(implicacion(True, False))
print(implicacion(False, True))
print(implicacion(False, False))
````

````python
#En este ejemplo se utilizaron ciclos, sin embargo eso no es lo importante sino la aplicación de la funcion para reutilizar un segmento de código.
def ord(list_:list):
  if (type(list_)==type([1,2,3])):
    for i in range(len(list_)):
      for j in range(i,len(list_)):
        if list_[i]>list_[j]:
          temp=list_[i]
          list_[i]=list_[j]
          list_[j]=temp
    return list_

a=[1,5,61,2,6,2,3,7,7]
print(ord(a))
b=[1,5,61,3,6,7,3,6,2]

print(ord(b))
````

````python
pi=3.14

def rad_to_grad(a):#en este caso permite leer pi porque esta dentro de un grupo de variables llamadas "globales"
  return a*(180/pi)


print(rad_to_grad(1*pi))
````

#Funciones incognitas o lambda

A parte de las funciones como segmentos de codigo tambien hay funciones que solo ejecutan una linea de codigo y que normalmente solo se utilizan para guardar funciones matematicas, estas son funciones lambda



```python
<Variable>=lambda <parametro 1>,<parametro 2>,...,<parametro n>: <linea de codigo>
```



**Ejemplo 1**

````python
f=lambda m: 0 if m<=0 else 1

#evaluando la función
for i in range(-10,10):
  print("f({})={}".format(i,f(i)))
````

**Ejemplo 2**

````python
g=lambda x,y: pow(x**2+y**2,1/2)

#evaluando la función
for i in range(-10,10,10):
  for j in range(-10,10,10):
    print("g({},{})={}".format(i,j,g(i,j)))
````
