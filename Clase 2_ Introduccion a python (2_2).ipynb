Los ciclos son estructuras que se repiten segun una condición

Untitled Diagram (2).png

Hay dos tipos de ciclos


Ciclo while, se repetira mientras se cumpla la condición
```python
while <Condicion>:
  <codigo...>
```

ciclo for, este tiene un funcionamiento diferente, puede iterar con una lista o con un rango, se detendrá antes de el ultimo valor del rango, es decir que si se coloca 10 terminará en 9
```python
for i in range(<numero de repeticiones>):
  <codigo...>
```

Tambien aceptan las sentencias `break` y `continue`, la primera detendrá el ciclo y el segundo hará que pase al siguiente paso del ciclo.

**Ejemplo 1**

factorial

$$n!=\prod_{m=1}^{n}m$$

Esta definición de factorial puede dar la idea de hacer un algoritmo iterativo


Untitled Diagram (4).png

````python
def factorial(a):
  valor=1
  for i in range(1,a+1):
    valor*=i
  return valor

i=0
while i<12:
  print("{:2}! = {}".format(i,factorial(i)))
  i+=1
````

uso de `break` y `continue` para interrumpir el ciclo o continuar el ciclo desde el siguiente

#try-except
Hay ocaciones que alguna operación no se puede cumplir porque genera un error, esto significa que el programa se detiene sin completar el proceso, para evitar esto se puede usar try para manejar estos errores como excepciones


```python
try:
  <codigo...>#Se ejecutara si no ocurre ningun error
except:
  <codigo...>#Se ejecuta si ocurre un error
```



**Ejemplo 1**

````python
for i in range(-10,10):
  try:
    print(1/i)
  except:
    print("idefinido")
````

**Ejemplo 2**

````python
a=None
try:
  a=open("archivo.txt","r")
  print("No hay error")
except:
  a=open("archivo.txt","w")
  a.write("contenido")
  a.close()
  a=open("archivo.txt","r")
  print("Hubo error")
finally:
  print(a.read())
````

#listas

Las listas son "arreglos" dinámicos que permiten añadir elementos de diferentes tipos y se puede añadir cuantos elementos se requieran sin definirlo al inicio.



**Ejemplo 1**

lista como iterador

````python
lista=["1",2,(3,4),5.6,True,[2,5],2+4j]

for i in lista:
  print(i,"\t\t\t",type(i))
````

Las listas, tuplas y cadenas son elementos a los cuales se puede acceder a sus subelementos

**Ejemplo 2**


````python
tupla=(2,3)
cadena="Ejemplo de cadena"
lista=[i for i in range(0,10,2)]# esta es una forma de definir una lista mediante un for, el cual asignará al valor previo una cantidad definida de veces, en este caso el valor es el iterador i

for t in [tupla, cadena, lista]:
  print(type(t))
  for i in t:
    print(i)
````

**Ejemplo 3**

ordenar una lista

````python
import random
def ordenar(x):
  for i in range(len(x)):
    for j in range(i,len(x)):
      if x[i]>x[j]:
        temp=x[i]
        x[i]=x[j]
        x[j]=temp
  return x

print(ordenar([ random.randint(0,20) for _ in range(10)]))
````

**Ejemplo 2**

````python
def suma(a,b):
  return a+b
def resta(a,b):
  return suma(a,-b)
def multi(a,b):
  return a*b
def division(a,b):
  return multi(a,1/b)

funcion={
    "suma":suma,
    "resta":resta,
    "multiplicacion":multi,
    "division":division
}


f=funcion.get("multiplicacion")
print(f(16,6))

f=funcion.get("suma")
f(16,6)
````

#Recursividad

Una funcion recursiva es una funcion que puede definirse a si misma, un ejemplo de ello es el factorial cuya definicion se da con el mismo
$$n!=n(n-1)!$$


Untitled Diagram (5).png

Como puede observarse **no se definen variables ni ciclos a excepcion del parámetro**


**Ejemplo 1**

````python
factorial=lambda n: (n*factorial(n-1) if n>0 else 1)

for i in range(12):
  print("{:2}!={}".format(i,factorial(i)))
````

**Ejemplo 2**

````python
def fibonachi(n,a=0,b=1):
  if n>0:
    return a+fibonachi(n-1,b,a+b)
  else:
    return 1

for i in range(12):
  print("fibonachi({}) = {}".format(i,fibonachi(i)))
````
